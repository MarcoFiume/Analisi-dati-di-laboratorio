```python
import numpy as np
import scipy as sp
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.DataFrame()
df['hv'] = np.array([1100., 1120., 1140., 1160., 1180., 1200., 1220., 1240., 1260., 1280., 1300., 1320., 1340., 1360., 1380., 1400.])
df['nd'] = np.array([103, 101, 101, 101, 101, 101, 101, 101, 102, 101, 101, 1002, 1101, 1101, 1101, 1101])
df['nt'] = np.array([2, 4, 2, 8, 15, 25, 38, 50, 58, 74, 77, 920, 1050, 1080, 1096, 1095])
df['eff']=df['nt']/df['nd']
df['err_eff']=np.sqrt(df['eff']*(1-df['eff'])/df['nd'])
```

```python
df
```

```python
fig, ax1 = plt.subplots(nrows=1, ncols = 1, figsize = (32,8))
ax1.grid(axis = 'y')
ax1.spines['right'].set_visible(False)
ax1.spines['top'].set_visible(False)
ax1.spines['bottom'].set_visible(False)
ax1.spines['left'].set_visible(False)
ax1.errorbar(df['hv'], df['eff'], df['err_eff'], fmt='o')
ax1.set_xlabel('Tensione di alimentazione (V)', fontsize = 22)
ax1.set_ylabel(r'$\eta_2$', fontsize = 22)
ax1.set_title('Efficienza', fontsize=22)
ax1.tick_params(axis='both', which='major', labelsize=14)
#plt.savefig('efficienza.svg', bbox_inches = "tight")
```
